```python

for i in range(len(x)):
    x.loc[i,'HTGS']=0
    x.loc[i,'ATGS']=0
    x.loc[i,'HTGC']=0
    x.loc[i,'ATGC']=0
    x.loc[i,'homeshots']=0
    x.loc[i,'awayshots']=0
    x.loc[i,'homeshotsT']=0
    x.loc[i,'awayshotsT']=0
    x.loc[i,'homefouls']=0
    x.loc[i,'awayfouls']=0
    x.loc[i,'homecorners']=0
    x.loc[i,'awaycorners']=0
    x.loc[i,'homeyellow']=0
    x.loc[i,'awayyellow']=0
    x.loc[i,'homered']=0
    x.loc[i,'awayred']=0
    HP=0
    AP=0
    a=0
    b=0
    c=0
    for j in range(i-1,-1,-1):
            if a<3:
                if x.loc[i,'HomeTeam']==x.loc[j,'HomeTeam']:
                    x.loc[i,'HTGS']=x.loc[j,'FTHG']+x.loc[i,'HTGS']
                    x.loc[i,'HTGC']=x.loc[j,'FTAG']+x.loc[i,'HTGC']
                    x.loc[i,'homeshots']=x.loc[j,'HS']+x.loc[i,'homeshots']
                    x.loc[i,'homeshotsT']=x.loc[j,'HST']+x.loc[i,'homeshotsT']
                    x.loc[i,'homefouls']=x.loc[j,'HF']+x.loc[i,'homefouls']
                    x.loc[i,'homecorners']=x.loc[j,'HC']+x.loc[i,'homecorners']
                    x.loc[i,'homeyellow']=x.loc[j,'HY']+x.loc[i,'homeyellow']
                    x.loc[i,'homered']=x.loc[j,'HR']+x.loc[i,'homered']
                    if x.loc[j,'FTR']=='H':
                        x.loc[i,'HM'+str(a+1)]='W'
                    elif x.loc[j,'FTR']=='A':
                        x.loc[i,'HM'+str(a+1)]='L'
                    else:
                        x.loc[i,'HM'+str(a+1)]='D'
                    a=a+1
                elif x.loc[i,'HomeTeam']==x.loc[j,'AwayTeam']:
                    x.loc[i,'HTGS']=x.loc[j,'FTAG']+x.loc[i,'HTGS']
                    x.loc[i,'HTGC']=x.loc[j,'FTHG']+x.loc[i,'HTGC']
                    x.loc[i,'homeshots']=x.loc[j,'AS']+x.loc[i,'homeshots']
                    x.loc[i,'homeshotsT']=x.loc[j,'AST']+x.loc[i,'homeshotsT']
                    x.loc[i,'homefouls']=x.loc[j,'AF']+x.loc[i,'homefouls']
                    x.loc[i,'homecorners']=x.loc[j,'AC']+x.loc[i,'homecorners']
                    x.loc[i,'homeyellow']=x.loc[j,'AY']+x.loc[i,'homeyellow']
                    x.loc[i,'homered']=x.loc[j,'AR']+x.loc[i,'homered']
                    if x.loc[j,'FTR']=='H':
                        x.loc[i,'HM'+str(a+1)]='L'
                    elif x.loc[j,'FTR']=='A':
                        x.loc[i,'HM'+str(a+1)]='W'
                    else:
                        x.loc[i,'HM'+str(a+1)]='D'
                    a=a+1
            if b<3:
                if x.loc[i,'AwayTeam']==x.loc[j,'AwayTeam']:
                    x.loc[i,'ATGS']=x.loc[j,'FTAG']+x.loc[i,'ATGS']
                    x.loc[i,'ATGC']=x.loc[j,'FTHG']+x.loc[i,'ATGC']
                    x.loc[i,'awayshots']=x.loc[j,'AS']+x.loc[i,'awayshots']
                    x.loc[i,'awayshotsT']=x.loc[j,'AST']+x.loc[i,'awayshotsT']
                    x.loc[i,'awayfouls']=x.loc[j,'AF']+x.loc[i,'awayfouls']
                    x.loc[i,'awaycorners']=x.loc[j,'AC']+x.loc[i,'awaycorners']
                    x.loc[i,'awayyellow']=x.loc[j,'AY']+x.loc[i,'awayyellow']
                    x.loc[i,'awayred']=x.loc[j,'AR']+x.loc[i,'awayred']
                    if x.loc[j,'FTR']=='H':
                        x.loc[i,'AM'+str(b+1)]='L'
                    elif x.loc[j,'FTR']=='A':
                        x.loc[i,'AM'+str(b+1)]='W'
                    else:
                        x.loc[i,'AM'+str(b+1)]='D'
                    b=b+1
                elif x.loc[i,'AwayTeam']==x.loc[j,'HomeTeam']:
                    x.loc[i,'ATGS']=x.loc[j,'FTHG']+x.loc[i,'ATGS']
                    x.loc[i,'ATGC']=x.loc[j,'FTAG']+x.loc[i,'ATGC']
                    x.loc[i,'awayshots']=x.loc[j,'HS']+x.loc[i,'awayshots']
                    x.loc[i,'awayshotsT']=x.loc[j,'HST']+x.loc[i,'awayshotsT']
                    x.loc[i,'awayfouls']=x.loc[j,'HF']+x.loc[i,'awayfouls']
                    x.loc[i,'awaycorners']=x.loc[j,'HC']+x.loc[i,'awaycorners']
                    x.loc[i,'awayyellow']=x.loc[j,'HY']+x.loc[i,'awayyellow']
                    x.loc[i,'awayred']=x.loc[j,'HR']+x.loc[i,'awayred']
                    if x.loc[j,'FTR']=='H':
                        x.loc[i,'AM'+str(b+1)]='W'
                    elif x.loc[j,'FTR']=='A':
                        x.loc[i,'AM'+str(b+1)]='L'
                    else:
                        x.loc[i,'AM'+str(b+1)]='D'
                    b=b+1
            if c<3:
                if x.loc[i,'HomeTeam']==x.loc[j,'HomeTeam'] and x.loc[i,'AwayTeam']==x.loc[j,'AwayTeam']:
                    if x.loc[j,'FTR']=='H':
                        HP+=3
                    elif x.loc[j,'FTR']=='A':
                        AP+=3
                    else:
                        HP+=1
                        AP+=1
                    x.loc[i,'HP']=HP
                    x.loc[i,'AP']=AP
                    c=c+1
                       
                    
                elif x.loc[i,'HomeTeam']==x.loc[j,'AwayTeam'] and x.loc[i,'AwayTeam']==x.loc[j,'HomeTeam']:
                    if x.loc[j,'FTR'] == 'H':
                        AP+=3
                    elif x.loc[j,'FTR']=='A':
                        HP+=3
                    else:
                        HP+=1
                        AP+=1
                    x.loc[i,'HP']=HP
                    x.loc[i,'AP']=AP
                    c=c+1
            
                
                
            
                
            
                    
                    
           
```

```python
def point_table(x):
    for i in range(0,380):
        x.loc[i,'HTP']=0
        x.loc[i,'ATP']=0
        for j in range(i-1,-1,-1):
            if x.loc[i,'HomeTeam']==x.loc[j,'HomeTeam']:
                if x.loc[j,'FTR']=='H':
                    x.loc[i,'HTP']=x.loc[i,'HTP']+3
                elif x.loc[j,'FTR']=='A':
                    x.loc[i,'HTP']=x.loc[i,'HTP']
                else:
                    x.loc[i,'HTP']=x.loc[i,'HTP']+1
            elif x.loc[i,'HomeTeam']==x.loc[j,'AwayTeam']:
                if x.loc[j,'FTR']=='H':
                    x.loc[i,'HTP']=x.loc[i,'HTP']
                elif x.loc[j,'FTR']=='A':
                    x.loc[i,'HTP']=x.loc[i,'HTP']+3
                else:
                    x.loc[i,'HTP']=x.loc[i,'HTP']+1
            if x.loc[i,'AwayTeam']==x.loc[j,'AwayTeam']:
                if x.loc[j,'FTR']=='H':
                    x.loc[i,'ATP']=x.loc[i,'ATP']
                elif x.loc[j,'FTR']=='A':
                    x.loc[i,'ATP']=x.loc[i,'ATP']+3
                else:
                    x.loc[i,'ATP']=x.loc[i,'ATP']+1
            elif x.loc[i,'AwayTeam']==x.loc[j,'HomeTeam']:
                if x.loc[j,'FTR']=='H':
                    x.loc[i,'ATP']=x.loc[i,'ATP']+3
                elif x.loc[j,'FTR']=='A':
                    x.loc[i,'ATP']=x.loc[i,'ATP']
                else:
                    x.loc[i,'ATP']=x.loc[i,'ATP']+1
    return x
                    
                    
```
